```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from adjustText import adjust_text
```

```python
# Importing play by play data 
df = pd.read_csv('PlayByPlay_Season22.csv')
df
```

```python
# Drop unnecessary columsn
df = df.loc[:, df.columns[3:]]
df
```

```python
df['scoreDiff'] = df['POINTS_A'] - df['POINTS_B']
```

```python
eventList = ['3FGM','2FGM','FTM']
```

```python
madeShots = df[df['PLAYTYPE'].isin(eventList)]
madeShots
```

# Clutch

```python
clutch = madeShots[madeShots['MINUTE'] >= 37]
clutch
```

```python
clutchPlayers = clutch.groupby(['PLAYER','PLAYTYPE'])['NUMBEROFPLAY'].count().reset_index()
clutchPlayers
```

```python
clutchPlayersPivoted = clutchPlayers.pivot(index='PLAYER', columns='PLAYTYPE', values='NUMBEROFPLAY')
clutchPlayersPivoted = clutchPlayersPivoted.fillna(0)
clutchPlayersPivoted['Clutch Points'] = clutchPlayersPivoted['2FGM'] * 2 +  clutchPlayersPivoted['3FGM'] *3 +  clutchPlayersPivoted['FTM']  
clutchPlayersPivoted.sort_values(by='Clutch Points',ascending = False).head(15)
```

# Late and Close

```python
lateClose = madeShots[madeShots['MINUTE'] >= 39]
lateClose = lateClose[(lateClose['scoreDiff'] >= -4) & (lateClose['scoreDiff'] <= 4)]
lateClose
```

```python
lateClosePlayers = lateClose.groupby(['PLAYER','PLAYTYPE'])['NUMBEROFPLAY'].count().reset_index()
lateClosePlayers
```

```python
lateClosePlayersPivoted = lateClosePlayers.pivot(index='PLAYER', columns='PLAYTYPE', values='NUMBEROFPLAY')
lateClosePlayersPivoted = lateClosePlayersPivoted.fillna(0)
lateClosePlayersPivoted['Clutch Points'] = lateClosePlayersPivoted['2FGM'] * 2 +  lateClosePlayersPivoted['3FGM'] *3 +  lateClosePlayersPivoted['FTM']  
lateClosePlayersPivoted.sort_values(by='Clutch Points',ascending = False).head(15)
```
